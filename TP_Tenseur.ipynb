```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color

```

```python
#charger l'image
image = data.camera()

image_float = image.astype(np.float64)
U, S, Vt    = np.linalg.svd(image_float) # calcul de svd

#affichage
plt.plot(S)
plt.title("valeur_singuliere_image")
plt.xlabel("index")
plt.ylabel("valeur singuliere ")
plt.show()
```

```python
k = 10
S_reduced     = np.zeros_like(S)
S_reduced[:k] = S[:k]

# reconstruction de l'image :
reconstructed_image = np.dot(U, np.dot(np.diag(S_reduced), Vt))

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("image_camera")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("imag_reconstruite k =10")
plt.imshow(reconstructed_image, cmap='gray')
plt.axis('off')

plt.show()

```
